```python
import numpy

science2k_doc_word = numpy.load("science2k-doc-word.npy")
print(science2k_doc_word.shape)
print(science2k_doc_word[:1])
```

```python
# We can visualize this dataset in 2D
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = [12, 4]

plt.scatter(science2k_doc_word[:,0], science2k_doc_word[:,1], alpha=0.5)
plt.ylabel("2nd word")
plt.xlabel("1st word")
plt.title("Dataset of Science 2000")
```

```python
# fit K-Means with K=4
from sklearn import cluster
model = cluster.KMeans(n_clusters=4)
model.fit(science2k_doc_word[:,[0,1]])
```

```python
# display the clusters in 2D
plt.scatter(science2k_doc_word[:,0], science2k_doc_word[:,1], alpha=0.5)
plt.scatter(model.cluster_centers_[:,0], model.cluster_centers_[:,1], marker='D', c='r', s=100)
plt.ylabel("2nd word")
plt.xlabel("1st word")
plt.title("Dataset of Science 2000")
plt.legend(['Datapoints', 'Probability peaks'])
```
